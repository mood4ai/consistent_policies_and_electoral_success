```python
!pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("white")
my_dpi=96
```

```python
df = pd.read_csv("data/parties_change.csv", sep=";", encoding = "ISO-8859-1")
df.tail(5)
```

```python
# transforming the catagorical column 
df['area']=pd.Categorical(df['area'])

for i in df.step.unique():
 
    
    fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)

    
    tmp=df[ df.step == i ]
    plt.scatter(tmp['change'], tmp['pop'] , s=tmp['pv']*40 , 
                c=tmp['area'].cat.codes, cmap="Accent", alpha=0.6, 
                edgecolors="white", linewidth=2)

    
    plt.yscale('linear')
    plt.xlabel("Change of politica convicions")
    plt.ylabel("Change in popularity")
    plt.title("Election: "+str(i) )
    plt.ylim(-300,300)
    plt.xlim(-20, 7000)

    
    filename='election_step'+str(i)+'.png'
    plt.savefig(filename, dpi=96)
    plt.gca()
```
